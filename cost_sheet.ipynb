```python
import matplotlib.pyplot as plt
```

```python
# Assumptions
price_per_sale = 250  # Average price of each sale in INR
sales_per_day = 75  # Number of sales per day
days_per_month = 30  # Operating days per month

# Costs
staff_salary_per_month = 15000 * 4  # 4 staff members with 15,000 INR per month salary
rent_per_month = 100000  # Rent cost per month in INR
utilities_per_month = 20000  # Utilities cost per month in INR
miscellaneous_per_month = 30000  # Miscellaneous expenses per month in INR
cost_of_ingredients_percentage = 0.35  # 35% of revenue goes towards ingredients
```

```python
# Revenue calculation
monthly_revenue = price_per_sale * sales_per_day * days_per_month

# Cost of goods sold (COGS) calculation
COGS = monthly_revenue * cost_of_ingredients_percentage

# Gross profit calculation
gross_profit = monthly_revenue - COGS

# Total expenses calculation
total_expenses = staff_salary_per_month + rent_per_month + utilities_per_month + miscellaneous_per_month

# Net profit calculation
net_profit = gross_profit - total_expenses

# Profit margin calculation
profit_margin = (net_profit / monthly_revenue) * 100

# Display results
print(f"Monthly Revenue: ₹{monthly_revenue}")
print(f"Cost of Goods Sold (COGS): ₹{COGS}")
print(f"Gross Profit: ₹{gross_profit}")
print(f"Total Expenses: ₹{total_expenses}")
print(f"Net Profit: ₹{net_profit}")
print(f"Profit Margin: {profit_margin:.2f}%")


# Data for plotting
labels = ['Revenue', 'COGS', 'Net Profit']
values = [monthly_revenue, COGS, net_profit]

expenses_labels = ['Staff Salaries', 'Rent', 'Utilities', 'Miscellaneous']
expenses_values = [staff_salary_per_month, rent_per_month, utilities_per_month, miscellaneous_per_month]

# 1. Bar Plot - Revenue vs COGS vs Net Profit
plt.figure(figsize=(6, 4))
plt.bar(labels, values, color=['#1f77b4', '#ff7f0e', '#9467bd'])
plt.title('Revenue vs COGS vs Net Profit', fontsize=14)
plt.ylabel('Amount in INR')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

# 2. Pie Chart - Expense Breakdown
plt.figure(figsize=(4, 4))
plt.pie(expenses_values, labels=expenses_labels, autopct='%1.1f%%', startangle=140, colors=['#1f77b4', '#ff7f0e', '#9467bd', '#ffdd44'])
plt.title('Expense Breakdown', fontsize=14)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# 3. Line Plot - Profit Margin Trend (using different sales per day)
sales_per_day_range = range(50, 101, 5)  # Simulate sales per day from 50 to 100
gross_profits = []

# Calculate gross profit for each sales per day
for sales in sales_per_day_range:
    revenue = price_per_sale * sales * days_per_month
    COGS = revenue * cost_of_ingredients_percentage
    gross_profit = revenue - COGS  # Gross profit is revenue minus COGS
    gross_profits.append(gross_profit)

plt.figure(figsize=(6, 4))
plt.plot(sales_per_day_range, gross_profits, marker='o', color='#9467bd', linestyle='-', linewidth=2)
plt.title('Profit Margin vs Sales per Day', fontsize=14)
plt.xlabel('Sales per Day')
plt.ylabel('Profit INR')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

```
